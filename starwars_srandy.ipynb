```python
import pulp as pl
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

import tabulky charakterů ze starwars do pandas dataframe
source - https://www.kaggle.com/datasets/jsphyg/star-wars?resource=download&select=characters.csv

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Source files/StarWars/characters.csv')
data[:10]
```

```python
agregated_data = data.groupby('species')['height'].mean().reset_index()
agregated_data = agregated_data.rename(columns={'height': 'avg height'})
```

```python
agregated_data.plot(x='species', y='avg height', title = 'Průměrná výška postav podle druhu', kind = 'bar')
plt.xlabel('Druh')
plt.xlabel('Průměrná výška (cm)')
plt.show()

agregated_data[:5]
```
