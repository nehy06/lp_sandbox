```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vstup = [1, 2, 3, 4, 5, 6]
produkce = [10, 25, 45, 60, 70, 75]
```

```python
prumerna_proudkce = []


for i in range(len(vstup)):
    prumerna_proudkce.append(vstup[i] / produkce[i])

prumerna_proudkce
```

```python
mezni_produkce = []

for i in range(1, len(vstup)):
    mezni_produkt = (produkce[i]-produkce[i-1])/(vstup[i]-vstup[i-1])
    mezni_produkce.append(mezni_produkt)

mezni_produkce
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))

ax1.plot(vstup, produkce, label='Celková produkce')
ax1.set_title('Celková produkce')
ax1.set_xlabel('Počet vstupů')
ax1.set_ylabel('Počet výstupů')
ax1.legend()

ax2.plot(vstup, prumerna_proudkce, label='Průměrná produkce')
ax2.set_title('Průměrná produkce')
ax2.set_xlabel('Počet vstupů')
ax2.set_ylabel('Průměrný počet výstupů')
ax2.legend()

ax3.plot(vstup[1:], mezni_produkce, label='Mezní produkce')
ax3.set_title('Mezní produkce')
ax3.set_xlabel('Počet vstupů')
ax3.set_ylabel('Mezní produkce')
ax3.legend()

plt.tight_layout()
plt.show()
```
